```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
webpage = requests.get('https://www.freelancer.in/jobs')
```

```python
soup = BeautifulSoup(webpage.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
job_listings = soup.find_all('div', class_='JobSearchCard-item')
job_listings
```

```python
data = []
for job in job_listings:
    job_title = job.find('a', class_='JobSearchCard-primary-heading-link').text.strip()
    job_description = job.find('p', class_='JobSearchCard-primary-description').text.strip()
    job_requirements = job.find('div', class_='JobSearchCard-primary-tags').text.strip()
    bid_price = job.find('div', class_='JobSearchCard-primary-price').text.strip()
    
    print("Job Title:", job_title)
    print("Description:", job_description)
    print("Requirements:", job_requirements)
    print("Bid Price:", bid_price)

    data.append({
        "Job Title": job_title,
        "Description": job_description,
        "Requirements": job_requirements,
        "Bid Price": bid_price
    })
```

```python
df = pd.DataFrame(data)
df
```
